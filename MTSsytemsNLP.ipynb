```python
from transformers import MarianMTModel, MarianTokenizer
```

```python
# Specifying the MarianMT model for English to Spanish translation.
model_name = "Helsinki-NLP/opus-mt-en-es"
```

```python

english_text = 'Give someone the benefit of the doubt'
```

```python

def translate(text):
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    translated_tokens = model.generate(**inputs)
    return tokenizer.decode(translated_tokens[0], skip_special_tokens=True)
```

```python

spanish_translation = translate(english_text)
```

```python

# Write the output to a file
output_filename = "translation_output.txt"
with open(output_filename, "w", encoding="utf-8") as file:
    file.write("Translation using MarianMT model:\n\n")
    file.write(f"English: {english_text}\n")
    file.write(f"Spanish: {spanish_translation}\n")
```

```python

print(f"Translation results saved to {output_filename}")
```
